```python
import gym
from stable_baselines3 import A2C
from stable_baselines3.common.env_util import make_vec_env

SEED = 44

env = make_vec_env("BipedalWalker-v3", n_envs=4, seed=SEED)

model = A2C("MlpPolicy", env, verbose=1, seed=SEED)
model.learn(total_timesteps=122000)
model.save("bipedal_walker")

```

```python
import gnwrapper

env = gnwrapper.Monitor(gym.make("BipedalWalker-v3",render_mode="rgb_array"),directory="./p/",)

observation = env.reset()
total_reward = 0
while True:
    action, _states = model.predict(observation,deterministic=True)
    observation, reward, done, info = env.step(action)
    total_reward += reward
    if done:
        env.reset()
        break

env.display()
print(env.videos)

print(f"Total reward: {total_reward}")
```
